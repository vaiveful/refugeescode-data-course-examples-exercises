# Choosing the Right Chart Type

The best chart depends on the question you are trying to answer! 

1. Comparison (Which is bigger?)
2. Trend over time (How does it change?)
3. Distribution (How is it spread?)
4. Relationship (How do variables relate?)
5. Composition (What makes up a whole?)


## 1. Comparison - Which is bigger?
- **Bar chart**
    - Use when: Comparing categories (sales by region, revenue by product)
    - Data: Categorical + numerical
    - Strength: Easy to read, familiar
    - Avoid when: Too many categories (becomes cluttered)
- **Horizontal bar**
    - Use when: Long category names or many categories
    - Why better: Labels are readable
    - Avoid when: Few categories (vertical is cleaner)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.DataFrame({
    "Country": ["Spain", "France", "Germany", "Italy"],
    "Revenue": [100, 120, 90, 110]
})

plt.figure()
sns.barplot(data=data, x="Country", y="Revenue")
plt.title("Revenue by Country")
plt.show()
```

```python
data = pd.DataFrame({
    "Product": [
        "Procrastinato 3000",
        "Emotional Support Water Bottle",
        "Socks of Uncertain Matching"
    ],
    "Sales": [200, 150, 180]
})

plt.figure()
sns.barplot(data=data, y="Product", x="Sales")
plt.title("Sales by Product")
plt.show()
```

## 2. Trend Over Time - How does it change?

- **Line chart**
    - Use when: Time series (monthly sales, daily users)
    - Data: Continuous (time)
    - Strength: Shows direction clearly
    - Avoid when: Data isn’t continuous (don’t fake trends)

- **Area chart**
    - Use when: Trend + magnitude emphasis
    - Strength: Visually impactful
    - Avoid when: Comparing multiple series (gets messy fast)

```python
import numpy as np

dates = pd.date_range(start="2024-01-01", periods=30)

data = pd.DataFrame({
    "Date": dates,
    "Users": np.random.randint(100, 200, size=30)
})

plt.figure()
sns.lineplot(data=data, x="Date", y="Users")
plt.title("Users Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.fill_between(data["Date"], data["Users"], alpha=0.3)
plt.plot(data["Date"], data["Users"])
plt.title("Users Over Time (Area)")
plt.xticks(rotation=45)
plt.show()
```

## 3. Distribution - How is it spread?

- **Histogram**
    - Use when: Frequency distribution (e.g. salaries)
    - Data: Continuous
    - Strength: Shows shape (normal, skewed)
    - Avoid when: Small dataset (misleading bins)
- **Box plot**
    - Use when: Comparing distributions across groups
    - Shows: Median, quartiles, outliers
    - Strength: Compact + analytical
    - Avoid when: Audience is non-technical (they get confused)
- **Violin plot**
    - Use when: You want distribution shape + density
    - Strength: More informative than box plot
    - Avoid when: Simplicity matters (can overwhelm)


```python
#Histogram
data = np.random.normal(loc=50, scale=10, size=500)

plt.figure()
sns.histplot(data, bins=30)
plt.title("Distribution of Values")
plt.show()

#Box Plot

df = pd.DataFrame({
    "Category": np.random.choice(["A", "B"], 200),
    "Value": np.random.normal(50, 15, 200)
})

plt.figure()
sns.boxplot(data=df, x="Category", y="Value")
plt.title("Box Plot by Category")
plt.show()

#Violin Plot
plt.figure()
sns.violinplot(data=df, x="Category", y="Value")
plt.title("Violin Plot by Category")
plt.show()
```

## 4. Relationship - How do variables relate?

- **Scatter plot**
    - Use when: Correlation (e.g. marketing spend vs revenue)
    - Data: Numerical vs numerical
    - Strength: Shows patterns, clusters, outliers
    - Avoid when: Too many overlapping points (use transparency)
- **Bubble chart**
    - Use when: Add third variable (size = meaning)
    - Strength: Multidimensional
    - Avoid when: Precision matters (humans are bad at comparing areas)



```python
#Scatter Plot
np.random.seed(42)

df = pd.DataFrame({
    "Ad_Spend": np.linspace(100, 1000, 100),
})
df["Revenue"] = df["Ad_Spend"] * 3 + np.random.normal(0, 200, 100)

plt.figure()
sns.scatterplot(data=df, x="Ad_Spend", y="Revenue")
plt.title("Ad Spend vs Revenue")
plt.show()


#Bubble Chart
df = pd.DataFrame({
    "Hours_Studied": np.random.uniform(0, 10, 100),
})

df["Score"] = 50 + df["Hours_Studied"] * 5 + np.random.normal(0, 15, 100)

plt.figure()
plt.scatter(
    df["Hours_Studied"],
    df["Score"],
    s=df["Score"],  # size
    alpha=0.5
)
plt.title("Bubble Chart: Hours Studied vs Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

## 5. Composition - What makes up a whole?

- **Stacked bar chart**
    - Use when: Part-to-whole across categories
    - Strength: Shows totals + breakdown
    - Avoid when: Too many segments (hard to compare)
- **100% stacked bar chart**
    - Use when: Comparing proportions
    - Strength: Normalized comparison
    - Avoid when: Absolute values matter
- **Pie chart**
    - Use when: Very simple part-to-whole (≤5 categories)
    - Strength: Intuitive
    - Avoid when: Anything complex (almost always replaceable with bar chart)

Hard truth: pie charts are overused and often inferior.

```python
#Stacked Bar Chart

data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar"],
    "Product_A": [30, 40, 35],
    "Product_B": [20, 25, 30]
})

data.set_index("Month").plot(kind="bar", stacked=True)
plt.title("Sales Composition by Month")
plt.show()

#100% Stacked Bar 

data_pct = data.set_index("Month")
data_pct = data_pct.div(data_pct.sum(axis=1), axis=0)

data_pct.plot(kind="bar", stacked=True)
plt.title("100% Stacked Bar")
plt.show()

#Pie Chart 
data = pd.Series([40, 35, 25], index=["A", "B", "C"])

plt.figure()
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.title("Simple Composition")
plt.show()
```

```python
# 1. You have exam scores for 200 students and you want to understand how the scores are distributed. which graph would you use? 

#data
scores = np.random.normal(70, 10, 200)

sns.histplot(scores, bins=20)
plt.show()
```

```python
# 2. Let's say marketing team wants to know if the higher spend lead to higher revenue, which chart  would you choose to explore the relationships?

#data
np.random.seed(42)
Ad_Spend = np.random.uniform(100, 1000, 100)
noise    = np.random.normal(0, 400, 100)
Revenue  = 3.8 * Ad_Spend + 800 + noise

df = pd.DataFrame({"Ad_Spend": Ad_Spend, "Revenue": Revenue})

sns.scatterplot(data=df, x="Ad_Spend", y="Revenue")
plt.show()
```

```python
# 3. You are given revenue by country and your boss wants to know which country has the highest revenue, choose the best chart to compare performance across countries.

data = pd.DataFrame({
    "Country": ["Spain", "France", "Germany", "Italy", "Portugal"],
    "Revenue": [120, 150, 130, 90, 70]
})

sns.barplot(data=data, x="Country", y="Revenue")
plt.show()
```

```python
#4. We want to show how each product contributes to total monthly sales, which graph would work here? 

data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product_A": [120, 135, 150, 170, 160, 180],
    "Product_B": [100, 115, 130, 140, 150, 155],
    "Product_C": [80, 90, 95, 110, 120, 130]
})

#Stacked Bar Chart
df = data.set_index("Month")
df.plot(kind="bar", stacked=True)
plt.title("Stacked Bar Chart: Product Contribution by Month")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()

#100% Bar Chart
df_pct = df.div(df.sum(axis=1), axis=0)

df_pct.plot(kind="bar", stacked=True)
plt.title("100% Stacked Bar Chart: Product Share by Month")
plt.ylabel("Proportion")
plt.xlabel("Month")
plt.ylim(0, 1)
plt.show()

#Pie Chart

total_sales = df.sum()

plt.figure()
plt.pie(
    total_sales,
    labels=total_sales.index,
    autopct="%1.1f%%"
)
plt.title("Total Sales Share by Product")
plt.show()
```

```python
# 5. You are tracking daily ice cream sales at a local ice cream shop over one week. The owner wants to understand how sales change over time to plan staffing and inventory.

np.random.seed(42)

dates = pd.date_range(start="2024-07-01", periods=7)

data = pd.DataFrame({
    "Date": dates,
    "Vanilla": np.random.randint(50, 120, size=7),
    "Chocolate": np.random.randint(40, 100, size=7)
})

plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x="Date", y="Vanilla", label="Vanilla")
sns.lineplot(data=data, x="Date", y="Chocolate", label="Chocolate")

plt.title("Daily Ice Cream Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Cones Sold")

plt.legend()

plt.show()
```
